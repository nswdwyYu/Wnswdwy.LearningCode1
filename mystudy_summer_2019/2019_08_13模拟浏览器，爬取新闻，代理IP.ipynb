## __爬取新闻__

```python
import urllib.request
import re
data = urllib.request.urlopen("http://news.sina.com.cn/").read()
data1 = data.decode("utf-8","ignore")
pat = 'href="(http://news.sina.com.cn/.*?)">'
allurl = re.compile(pat).findall(data1)
for i in range(0,len(allurl)):
    try:
        print("这是第"+str(i)+"次爬取！！")
        thisurl = allurl[i]
        file = "F://Python_All/Python_Practice_File/sina_news/"+str(i)+".html"
        urllib.request.urlretrieve(thisurl,file)
        print("--爬取成功---")
        print("\n")
    except urllib.error.URLError as e:
        if hasattr(e,"code"):
            print(e.code)
        if hasattr(e,"reason"):
            print(e.reason)
        print("\n")
```

## __模拟浏览器爬取CSDN首页所有文章__

```python
import urllib.request
import re
url = "https://blog.csdn.net/"
headers = ("User-Agent","Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko)Chrome/76.0.3809.100 Safari/537.36")
opener = urllib.request.build_opener()
opener.addheaders = [headers]
urllib.request.install_opener(opener)
data = urllib.request.urlopen(url).read().decode("utf-8","ignore")
pat = '<h2>\n<a href="(.*?)"'
result = re.compile(pat).findall(data)
for i in range(0,len(result)):
    file = "F:/Python_All/Python_Practice_File/csdn_passage/"+str(i)+".html"
    urllib.request.urlretrieve(result[i+1],filename=file)
    print("第"+str(i)+"次爬虫成功")
```

## __使用代理IP爬取内容__

```python
import urllib.request
def use_proxy(url,proxy_add):
    proxy = urllib.request.ProxyHandler({"http":proxy_add})
    opener = urllib.request.build_opener(proxy,urllib.request.HTTPHandler)
    urllib.request.install_opener(opener)
    data = urllib.request.urlopen(url).read().decode("utf-8","ignore")
    return data
proxy_add = "119.183.220.224:8888"
url = "http://www.baidu.com"
data = use_proxy(url,proxy_add)
print(len(data))
```
