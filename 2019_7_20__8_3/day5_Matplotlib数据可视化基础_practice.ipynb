```python
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams['lines.linestyle']='-' #取值可为‘-’ '-.' '--'
plt.rcParams['lines.linewidth']=2   #取值为0-10  默认1.5
plt.rcParams['font.sans-serif']='SimHei'  #中文字体正常显示
plt.rcParams['axes.unicode_minus'] = False  #设置特殊符号正常显示

data2=np.arange(0,np.pi*2,0.01)
#第一步 绘制大画布
p_big=plt.figure(figsize=(11,10),dpi=200)
#第二步   绘制第1个子画布
ax_fist=p_big.add_subplot(2,2,1)
    #设置标题
plt.title('测试的图标')
plt.xlabel('X轴')
plt.ylabel('Y轴')
plt.plot(data2,data2**2)  # 在画布上画图
#绘制第2个子画布
ax_second=p_big.add_subplot(2,2,2)
plt.plot(data2,data2*np.pi)   #在画布上画图
#设置第3个子画布
ax_thrid = p_big.add_subplot(2,2,3)


#设置标题
plt.title('sin 的测设图表')
plt.xlabel('X轴')
plt.ylabel('Y轴')
#设置x/y的取值范围
plt.ylim(-1,1)
plt.xlim(0,np.pi*2)
#x/y轴的刻度范围
plt.yticks([-1,-0.5,0,0.5,1])
plt.xticks([0,np.pi/2,np.pi,np.pi*1.5,np.pi*2])
#绘图
plt.plot(data2,np.sin(data2))
#绘制图例
plt.legend(['sin'])
#绘制第4个子画布
ax_fourth=p_big.add_subplot(2,2,4)
plt.plot(data2,np.cos(data2))    #在画布上画图
plt.plot(['cos'])
#保存并显示
plt.savefig('F:\Python_All\Python_Practice_File/first1.jpg')
plt.show()
```
