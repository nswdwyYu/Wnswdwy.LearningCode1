### 解析URL库  -requests


```python
import requests
url = 'http://www.163.com'
ual={'Use-Agent':'User-Agent:Mozilla/5.0 (Windows; U; Windows NT 6.1; en-us) AppleWebKit/534.50 (KHTML, like Gecko)Version/5.1 Safari/534.50'}
resp=requests.get(url,headers=ual,timeout=3)
print(resp.status_code)
#print(resp.text)
```

```python
print(resp.encoding)
```

```python
#网页编码的修改

import chardet    #智能识别网站编码
print(chardet.detect(resp.content))

resp.encoding = chardet.detect(resp.content)['encoding']
print(resp.encoding)
```

## 解析网页
### 1.打开开发者工具
### 2.

```python
# 创建BeauSoup对象
from bs4 import BeautifulSoup
html = resp.content
soup = BeautifulSoup(html,'html.parser') #创建对象
soup.prettify(encoding='utf-8')

```

### 提取元素

```python
title = soup.find_all('title')
print(title[0].get_text())
```

```python
a=soup.find_all('table')
print(a)
```

```python
# 提取元素
title=soup('title')
print(title)
print(type(title))
print(type(title[0]))
print('tag对象',title[0])
print(type(title[0].string))
```

```python
title1 = soup.find('a')
print(title1)
```

```python
ul = soup.select('ul.cm_ul_round')
ul1=ul[0]
li = ul1.select('li')  #提取ul中的li标签
#print(li)
text_list=[]
href_list=[]

#把li标签下a标签的文本和链接href属性提取出来
for i in range(len(li)):
    a  =li[i].select('a')
    text_list.append(a[0].get_text())
    href_list.append(a[0].get('href'))
print(text_list)
print(href_list)
```

```python
import pandas as pd
dic1={}
dic1['text_list']=text_list
dic1['href_list']=href_list
df1=pd.DataFrame(dic1)
print(df1)
```
