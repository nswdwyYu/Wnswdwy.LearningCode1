```python
#箱线图
####固定的输入
import numpy as np
import matplotlib.pyplot as plt

#导入数据
data = np.load('C:/Users/nswdwy_yu/Desktop/国民经济核算季度数据.npz')
names=data['columns']
values=data['values']
#设置参数
plt.rcParams['font.sans-serif'] = 'SimHei'
plt.rcParams['axes.unicode_minus'] = False
#绘制画布
plt.figure(figsize = (10,8) ,dpi = 80)

#设置图表的各内容
plt.title('2000-2017年国民经济核算图表')
gdp_values=(list(values[:,3]),list(values[:,4]),list(values[:,5]))
plt.xlabel('产业')
plt.ylabel('产业值')
#plt.xlim()
plt.ylim(0,140000)
#plt.xticks()
plt.yticks(range(0,140000,20000))
#画图
label=['第1产业','第2产业','第3产业']
plt.boxplot(gdp_values,notch=True,labels=label)
#plt.scatter()
#图例
#plt.legend()
#保存并显示
plt.savefig('F:\Python_All\Python_Practice_File\catter.jpg')
plt.show()
```
