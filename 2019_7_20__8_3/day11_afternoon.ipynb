```python
###读取数据和简单的数据探索
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt

from sklearn import datasets
iris = datasets.load_iris()

iris.keys()
#print(iris.DESCR)#数据描述
iris.data.shape
iris.feature_names

X = iris.data[:,:2]
X.shape
plt.scatter(X[:,0],X[:,1])
plt.show()
y = iris.target
plt.scatter(X[y==0,0],X[y==0,1] ,color='red')#y所在0行0列和0行1列
plt.scatter(X[y==1,0],X[y==1,1] ,color='blue',marker='o')#marker设置不同样式
plt.scatter(X[y==2,0],X[y==2,1] ,color='green',marker='+')
plt.show()
```

## P2P网贷数据读取


```python
def dropNullStd(data):
    beforelen=data.shape[1]
    colisNull = data.describe().loc['count']==0
    for i in range(len(colisNull)):
        if colisNull[i]:
            data.drop(colisNull.index[i],axis=1,inplace=True)
    stdisZero=data.describe().loc['std']==0
    for i in range(len(stdisZero)):
        if stdisZero[i]:
            data.drop(stdisZero.index[i],axis=1,inplace=True)
    affterlen=data.shape[1]
    print('去除的列的数目为：',beforelen-affterlen)
    print('去除后数据的形状为：',data.shape)

```

```python
dropNullStd(master)
```
